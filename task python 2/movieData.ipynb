```python
data = pd.read_csv('u.data',delimiter='\t',names=["usr_id", "mv_id", "rating", "data"])
```

```python
userdata = pd.read_csv('u.user',delimiter='|',names=["usr_id", "age", "sex", "profession" ,"zip_code"])
```

```python
headerlist =[
    "mv_id",
    "title",
    "release_date",
    "video_release_date",
    "IMDb URL",
    "unknown",
    "Action",
    "Adventure",
    "Animation",
    "Childrens",
    "Comedy",
    "Crime",
    "Documentary",
    "Drama",
    "Fantasy",
    "Film-Noir",
    "Horror",
    "Musical",
    "Mystery",
    "Romance",
    "Sci-Fi",
    "Thriller",
    "War",
    "Western"
]
h2 =[
    "mv_id",
    "title",
    "release_date",
    "video_release_date",
    "IMDb URL"]
```

```python
mvData = pd.read_csv('u.item',sep='|', names=h2,encoding='latin-1',usecols=range(5))
```

```python
mvData.head(5)
```

```python
mvData['release_date'] =  pd.to_datetime(mvData['release_date'])
```

```python
mvData.info()
```

```python
mvRating = data.groupby(['mv_id']).count()
mvRating[['rating']].head()
```

```python
#top ten moves with highest ratings
mvRating = mvRating[['rating']].sort_values(['rating'],ascending=False)[:10]
```

```python
mvRating
```

```python
mvJoin = pd.merge(data,mvData,on ='mv_id')
```

```python
mvJoin = mvJoin.drop(columns=['data'])
mvJoin.head()
```

```python
mindate = mvJoin['release_date'].min()
```

```python
oldRated = mvJoin[(mvJoin['rating']==5) & (mvJoin['release_date'] == mindate)]
```

```python
#oldest released movie with top rating
#oldRated=oldRated.drop(columns=['usr_id'])
#oldRated.set_index('mv_id')
oldRated.drop_duplicates()
```

```python
userJoin = pd.merge(userdata,mvJoin,on ='usr_id')
```

```python
userJoin.head()
```

```python
mostratd = userJoin[(userJoin['profession']=='student')].groupby('title').count().sort_values(['rating'],ascending=False)[:1]
#most rated movie by students
mostratd[['rating']]
```

```python
ageRatd = userJoin[(userJoin['age']>= 20) & (userJoin['age']<= 25)].groupby('title').count().sort_values(['rating'],ascending=False)[:1]
#most rated movie by ages between 20-25
ageRatd[['rating']]
```

```python
#top movies rated 5 
mstratd = mvJoin[(mvJoin['rating']==5)].groupby('title').count().sort_values(['rating'],ascending=False)
```

```python
mstratd[['rating']].head()
```

```python
mstratdzipcode = userJoin.groupby('zip_code').count().sort_values(['rating'],ascending=False)[:10]
```

```python
# 10 zip code with most rating
mstratdzipcode[['rating']].head()
```
